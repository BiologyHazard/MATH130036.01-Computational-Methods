```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 抛物线均方逼近

def quadratic_approximation(x, y):
    A = x.reshape(-1, 1) ** np.arange(3)
    alpha = np.linalg.solve(A.T @ A, A.T @ y)
    return np.poly1d(alpha[::-1])
```

```python
x = np.arange(1, 12+1)
y = np.array([256, 201, 159, 61, 77, 40, 17, 25, 103, 156, 222, 345])
```

```python
approximation = quadratic_approximation(x, y)
print(approximation)
```

```python
plt.grid(which='both')
plt.plot(x, y, 'x', label='nodes')
x = np.linspace(1, 12, 1000+1)
plt.plot(x, approximation(x), label="approximation")
plt.legend()
plt.show()
```
