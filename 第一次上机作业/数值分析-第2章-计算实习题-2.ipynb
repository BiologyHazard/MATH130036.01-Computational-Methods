```python
import matplotlib.pyplot as plt
import numpy as np

from interpolation import make_cubic_spline_interpolation, make_polynomial_interpolation_equal_step
```

```python
def runge(x):
    return 1 / (1 + 25 * x**2)
```

```python
n = 10
start, stop = -1, 1
x = np.linspace(start, stop, n+1)
y = runge(x)
p = make_polynomial_interpolation_equal_step(start, stop, n, y)
s = make_cubic_spline_interpolation(x, y)

plt.grid()
plt.plot(x, y, 'x', label='nodes')
x = np.linspace(start, stop, 1000+1)
plt.plot(x, runge(x), label='original')
plt.plot(x, p(x), label='polynomial interpolation')
plt.plot(x, s(x), label='cubic spline (natural)')
plt.legend()
plt.show()
```
